```python
#string to csv\
import pandas as pd
info = '''Name, Address, Age\n Nischal, Kirtipur, 23\nHari, Lagankhel, 45'''
file = open('str_to_csv.csv', 'w')
file.write(info)
file.close()
```

```python
#list to csv

info = [['Name', 'Address', 'Age'],
       ['Hari', 'India', 23],
       ['Rabin', 'Balkumari', 24],
       ['Bimisha', 'Nakhipot', 25],
       ['Manas', 'Dharmasthali', 23]]
file = open('list_to_csv.csv', 'w')
import csv
x = csv.writer(file)
x.writerows(info)
file.close()
df = pd.read_csv('list_to_csv.csv')
df
```

```python
#dict to csv
info = {'name': ['Nischal', 'Rabin', 'Kaushal', 'Anish'],
       'Age': [23, 24, 22, 23],
       'Address':['Kirtipur', 'Balkumari', 'Kirtipur', 'le-le']}
df = pd.DataFrame(info)
df.to_csv('dict_to_csv.csv')
df
```
