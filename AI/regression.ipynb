```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('student_scores.csv',)
dataset = dataset.dropna() #drop rows having empty value or NAN value
dataset
```

```python
# dataset[['Years of Experience']]
```

```python
# dataset[['Years of Experience']].values
```

```python
X =  dataset.iloc[:, 0:-1].values
X
```

```python
Y =   dataset.iloc[:,1:].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state = 1)
```

```python
print(type(X_train[0][0]))
```

```python
# Training model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
exp = float(input("Enter the hours of study:"))
exp = np.array([[exp]])
print(exp)

#predict the salary
predicted_salary = model.predict(exp)[0][0]
print("Your predicted score is = ", predicted_salary)
```

```python
predicted_score = model.predict(X_test)
predicted_score
```

```python
plt.scatter(X_train, Y_train, color = 'blue')
plt.plot(X_train, model.predict(X_train), color = 'red')
plt.show()
```
