```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_data.csv')
df
```

```python
data = df[['Age', 'Education Level', 'Years of Experience', 'Salary']]
data
```

```python
data.isna().sum()
```

```python
data = data.dropna()
data
data.isna().sum()
# X = data.iloc[:, 0:1].values
# y = data.iloc[:, 1:2].values
```

```python
education = pd.get_dummies(data['Education Level'], drop_first = False)
education
```

```python
data = pd.concat([data, education], axis = 1)
data
```

```python
data = data.drop(['Education Level'], axis = 1)
data
```

```python
data = data[['Years of Experience', 'Salary']]
data
```

```python
X = data.iloc[:, 0:1].values
y = data.iloc[:, 1:2].values
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
train_x, df_test = train_test_split(data, train_size = 0.85, test_size = 0.15, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
# X_train = df_train[['Age','Years of Experience',"Bachelor's","Master's","PhD"]]
# y_train = df_train['Salary']
lr = LinearRegression()
lr.fit(X, y)
```

```python
plt.scatter(X, y, color='cyan')
plt.plot(X, lr.predict(X), color='magenta')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
