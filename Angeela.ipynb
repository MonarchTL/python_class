```python
def main_input():
    try:
        a = int(input("Enter the first number = "))
        o = input("Enter the operator: +,-,/,* = ")
        b = int(input("Enter the second number = "))
    except Exception as e:
        print(f"Error = {e}")
        main_input()
    result = calc(a, o, b)
    print(a, o, b, "=", result)
    ask_more(result)
def calc(a, o, b):
    try:
        if o == "+":
            return a + b
        elif o == "-":
            return a - b
        elif o == "/":
            return a / b 
        elif o == "*":
            return a * b
        else:
            print("Invalid operator...try again")
            main_input()
            
            
    except Exception as e:
        print(f"Error = {e}")
        main_input()
 


def ask_more(result):
    x = input("Do you want to continue? Y or N")
    if x == "Y":
        a = result
        cont(a)
        
    else:
        print("Done....")
        

def cont(a):
    o = input("Enter the operator: +,-,/,* = ")
    b = int(input("Enter the second number = "))
    print(f"a = {a}")
    result = calc(a, o, b)
    print (a, o, b, "=", result)
    ask_more(result)
    
main_input()
    
```

```python
#Add
def add(n1, n2):
  return n1 + n2
#Subtract
def sub(n1, n2):
  return n1 - n2
#division
def div(n1, n2):
  return n1 / n2
#multiplication
def mul(n1, n2):
  return n1 * n2
def calculator():
    n1 = int(input("Enter the first number = "))
    operations = {'+':add,'-':sub, '*':mul, '/':div}
    for items in operations:
      print(items)
    should_cont = True
    while should_cont == True:
        operator = input("Enter the operator:")
        n2 = int(input("Enter the next number number = "))

        operation_function = operations[operator]
        answer = operation_function(n1, n2)
        print(n1, operator, n2, "=", answer)
        ask_user = input (f"Do you want to continue with {answer} 'Y' or 'N' for new.:")
        if ask_user == 'Y':
            n1 = answer
            should_cont = True
        else:
            should_cont = False
            calculator()
calculator()
    

```
