```python
import requests
from bs4 import BeautifulSoup


url = BeautifulSoup("https://www.sharesansar.com/live-trading", 'html.parser')
soup = requests.get(url)


```

```python
code = BeautifulSoup(soup.text, 'lxml')
data = []
cleaned_data = []
# print(code)
for tags in code:
    new_text = tags.getText().split('\n')
    data.append(new_text)

cleaned_sublist = []
for sublist in data:    
    for s in sublist:
        if s.strip():
            cleaned_string = s.strip()
            cleaned_sublist.append(cleaned_string)

print(cleaned_sublist)
```
